```python
import random

CHOICES = [
    "apples",
    "bananas",
    "cherries",
    "dates",
    "elderberries",
    "figs",
    "grapes",
    "huckleberries",
    "imbers",
    "juneberries",
    "kiwis",
]


def generate_long_results():
    for _ in range(1000):
        for y in range(5):
            print(random.choice(CHOICES), end=" ")
        print()


generate_long_results()
```
